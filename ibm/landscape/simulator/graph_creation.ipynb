```python
import networkx as nx 
import matplotlib.pyplot as plt
import random
```

```python
number_of_nodes = 8
degree_of_each_node = 3
```

```python
G = nx.generators.random_graphs.random_regular_graph(degree_of_each_node, number_of_nodes, 42)
```

```python
nx.draw(G)
```

```python
def write_to_file(G):
    name = f"graph_3_reg_{len(G.nodes())}_{len(G.edges())}_w_01.txt"
    f = open(name, 'w')
    f.write(f"{len(G.nodes())}\n")
    for (u,v,w) in G.edges(data=True):
        f.write(f"{u+1},{v+1},{w['weight']}\n")
    f.close()
```

```python
for i in range(4, 21, 2):
    G = nx.generators.random_graphs.random_regular_graph(degree_of_each_node, i, 42)
    for (u,v,w) in G.edges(data=True):
        w['weight'] = random.randint(1,10)
    plt.figure()
    print(G.edges(data=True))
    nx.draw(G)
    plt.show()
    write_to_file(G)
```
